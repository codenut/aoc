```python
with open("inputs/day3.in") as f:
    engine = [line.strip() + "." for line in f.readlines()]
```

```python
from collections import defaultdict


def adj(row, col):
    return (
        (row, col + 1), (row, col - 1), (row + 1, col), (row - 1, col),
        (row + 1, col + 1), (row + 1, col - 1), (row - 1, col - 1), (row - 1, col + 1)
    )

def is_sym(c):
    return c != '.' and not c.isdigit()

def get_partnum():
    n, m = len(engine), len(engine[0])
    num, has_sym = 0, False
    nums = []
    gears = defaultdict(list)
    syms = []

    for i, row in enumerate(engine):
        for j, _ in enumerate(row):
            if engine[i][j].isdigit():
                for nrow, ncol in adj(i, j):
                    if 0 <= nrow < n and 0 <= ncol < m and is_sym(engine[nrow][ncol]):
                        has_sym = True
                        if engine[nrow][ncol] == '*':
                            syms.append((nrow, ncol))
                            
                num = num * 10 + int(engine[i][j])
            else:
                if has_sym:
                    nums.append(num)
                for sym in syms:
                    gears[sym].append(num)
                has_sym = False
                num = 0

    print(gears)
    return nums
```
